```python
X = df_cancer.drop(['target'], axis = 1) # We drop our "target" feature and use all the remaining features in our dataframe to train the model.
X.head()
```

```python
y = df_cancer['target']
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 20)
```

```python
print ('The size of our training "X" (input features) is', X_train.shape)
print ('\n')
print ('The size of our testing "X" (input features) is', X_test.shape)
print ('\n')
print ('The size of our training "y" (output feature) is', y_train.shape)
print ('\n')
print ('The size of our testing "y" (output features) is', y_test.shape)
```

```python
from sklearn.svm import SVC
svc_model = SVC()
svc_model.fit(X_train, y_train)
```

```python
y_predict = svc_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
cm = np.array(confusion_matrix(y_test, y_predict, labels=[1,0]))
confusion = pd.DataFrame(cm, index=['is_cancer', 'is_healthy'],
                         columns=['predicted_cancer','predicted_healthy'])
confusion
```

```python
sns.heatmap(confusion, annot=True)
```

```python
print(classification_report(y_test, y_predict))
```

Precision: Precision measures the proportion of predicted positive cases that were actually positive. It is calculated as:

Precision = TP / (TP + FP)

where TP is the number of true positives, and FP is the number of false positives.

Recall: Recall measures the proportion of actual positive cases that were correctly identified by the model. It is calculated as:

Recall = TP / (TP + FN)

where FN is the number of false negatives.

F1 score: F1 score is the harmonic mean of precision and recall, and provides a balanced measure of the two metrics. It is calculated as:

F1 score = 2 * Precision * Recall / (Precision + Recall)

Support: Support is simply the number of observations in each class. It can be used to identify imbalanced classes in the data, which can affect the performance of the model.

These metrics can be calculated for each class in a multi-class classification problem by taking a weighted average of the values for each class, or by calculating them separately for each class and then averaging them.

Note that in some cases, precision and recall may be more appropriate metrics than accuracy for evaluating the performance of a classification model. This is especially true when the classes are imbalanced, as accuracy can be misleading in such cases.

Normalize Training Data to improve model

```python
X_train_min = X_train.min()
X_train_max = X_train.max()
X_train_range = (X_train_max- X_train_min)
X_train_scaled = (X_train - X_train_min)/(X_train_range)
X_train_scaled.head()
```

```python
X_test_min = X_test.min()
X_test_range = (X_test - X_test_min).max()
X_test_scaled = (X_test - X_test_min)/X_test_range
svc_model = SVC()
svc_model.fit(X_train_scaled, y_train)
```

```python
y_predict = svc_model.predict(X_test_scaled)
cm = confusion_matrix(y_test, y_predict)
```

```python
cm = np.array(confusion_matrix(y_test, y_predict, labels=[1,0]))
confusion = pd.DataFrame(cm, index=['is_cancer', 'is_healthy'],
                         columns=['predicted_cancer','predicted_healthy'])
confusion
```

```python
sns.heatmap(confusion,annot=True)
```

```python
print(classification_report(y_test,y_predict))
```

The macro avg is calculated as the average of the precision, recall, and F1 score for each class separately. In other words, we calculate the precision, recall, and F1 score for each class (positive and negative) and then take the average of these scores. The macro avg treats each class equally and is useful when the class distribution is balanced.

The weighted avg, on the other hand, is calculated as the weighted average of the precision, recall, and F1 score for each class, weighted by the number of samples in each class. In other words, we calculate the precision, recall, and F1 score for each class (positive and negative) and then take the weighted average of these scores, where the weights are proportional to the number of samples in each class. The weighted avg is useful when the class distribution is imbalanced, as it gives more weight to the class with more samples.

In your example, the accuracy of the model on the test dataset is 0.96, which means that the model correctly predicted the class of 96% of the samples. The macro avg precision, recall, and F1 score are all 0.95, which indicates that the model performs similarly on both classes. The weighted avg precision, recall, and F1 score are also 0.96, which suggests that the model is robust to imbalanced class distribution and performs well on both classes, despite the fact that one class may have more samples than the other.

```python
# Further Model Improvement
# ** The search for the optimal set of hyperparameters is called gridsearching.**

# Gridsearching gets its name from the fact that we are searching over a "grid" of parameters. For example, imagine the alpha hyperparameters on the x-axis and fit_intercept on the y-axis, and we need to test all points on the grid.


# Searching for the best hyperparameters
# Let's see if we can improve on our model by searching for the best hyperparameters.

# We would need to evaluate on the training data the set of hyperparameters that perform best, and then use this set of hyperparameters to fit the final model and score on the testing set.

# Gridsearch Model
param_grid = {'C': [0.1, 1, 10, 100], 'gamma': [1, 0.1, 0.01, 0.001], 'kernel': ['rbf']} 
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=4)
grid.fit(X_train_scaled,y_train)
```

This line of code initializes a GridSearchCV object, which is a tool in scikit-learn used for hyperparameter tuning.

The first argument, SVC(), specifies the estimator or model to be optimized. In this case, it is the Support Vector Classifier (SVC).

The second argument, param_grid, is a dictionary or a list of dictionaries that contains the hyperparameters and their corresponding values to be optimized. Each key in param_grid corresponds to a hyperparameter name, and its corresponding value is a list of values to be tested. GridSearchCV will perform a search across all possible combinations of hyperparameters and their values to find the best combination that maximizes the performance metric.

The third argument, refit=True, means that the model will be refit with the best parameters found by GridSearchCV on the entire dataset after cross-validation.

The last argument, verbose=4, controls the level of detail of the log messages during the hyperparameter tuning process. verbose=4 is the highest level of verbosity and will print detailed messages for each candidate combination of hyperparameters.

estimator=SVC(): Specifies the estimator or model to be optimized. In this case, it is the Support Vector Classifier (SVC).

param_grid={'C': [0.1, 1, 10, 100], 'gamma': [1, 0.1, 0.01, 0.001], 'kernel': ['rbf']}: Specifies the hyperparameters and their corresponding values to be optimized. The dictionary param_grid has three keys: 'C', 'gamma', and 'kernel', and their corresponding values are lists of values to be tested for each hyperparameter. The hyperparameter 'C' controls the regularization strength, while 'gamma' controls the kernel coefficient. The 'kernel' hyperparameter specifies the type of kernel to be used. In this case, the only kernel being tested is the radial basis function kernel ('rbf').

verbose=4: Specifies the level of detail of the log messages during the hyperparameter tuning process. verbose=4 is the highest level of verbosity and will print detailed messages for each candidate combination of hyperparameters.

```python
print (grid.best_params_)
print ('\n')
print (grid.best_estimator_)
```

```python
grid_predictions = grid.predict(X_test_scaled)
cm = np.array(confusion_matrix(y_test, grid_predictions, labels=[1,0]))
confusion = pd.DataFrame(cm, index=['is_cancer', 'is_healthy'],
                         columns=['predicted_cancer','predicted_healthy'])
confusion
```

```python
print(classification_report(y_test,grid_predictions))
```
